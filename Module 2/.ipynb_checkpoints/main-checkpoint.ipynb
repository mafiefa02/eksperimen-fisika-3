# **Percobaan 1 - *Spin Coating***

## ***Visualization Setup***

### **Definisi Konstanta / Variasi**

```python
vol = 1e-4 # litre
t = 20 # sekon
rpm = [500, 1000, 2000, 3000, 4000] # RPM
```

### **Data Frames**

#### **Data P3HT**

```python
# Ketebalan (meter)
p3ht = {"Kecepatan sudut putar": rpm, "Ketebalan": [200, 100, 50, 34, 25]}
df_p3ht = pd.DataFrame(data=p3ht)

# Ganti RPM menjadi rad/s
df_p3ht["Kecepatan sudut putar"] = df_p3ht["Kecepatan sudut putar"].apply(lambda x: x * (2*np.pi/60))

# Ganti nanometer ke meter
df_p3ht['Ketebalan'] = df_p3ht['Ketebalan'].apply(lambda x: x/(1e+9))

df_p3ht.head()
```

```python
# Define constants
h_0_p3ht = max(df_p3ht['Ketebalan'])
c_0_p3ht = 5 # kg/m^3
x_p3ht = df_p3ht["Kecepatan sudut putar"]
y_p3ht = df_p3ht['Ketebalan']
x_fit_p3ht = np.linspace(min(df_p3ht["Kecepatan sudut putar"])/2, 3 * max(df_p3ht["Kecepatan sudut putar"])/2, 1000)
```

#### **Data R-Dye**

```python
# Ketebalan (meter)
r_dye = {'RPM': rpm, "Ketebalan": [65, 35, 15, 10, 8]}
df_r_dye = pd.DataFrame(data=r_dye)

# Ganti RPM menjadi rad/s
df_r_dye['RPM'] = df_r_dye['RPM'].apply(lambda x: x * (2*np.pi/60))

# Ganti nanometer ke meter
df_r_dye['Ketebalan'] = df_r_dye['Ketebalan'].apply(lambda x: x/(1e+9))

df_r_dye.head()
```

```python
# Define constants
h_0_r_dye = df_r_dye['Ketebalan'].iloc[0]
c_0_r_dye = 2 # kg/m^3
x_r_dye = df_r_dye['RPM']
y_r_dye = df_r_dye['Ketebalan']
x_fit_r_dye = np.linspace(df_r_dye['RPM'].iloc[0], df_r_dye['RPM'].iloc[-1], 1000)
```

#### **Data MEH-PPV**

```python
# Ketebalan (meter)
meh_ppv = {'RPM': rpm, "Ketebalan": [50, 20, 10, 5, 2]}
df_meh_ppv = pd.DataFrame(data=meh_ppv)

# Ganti RPM menjadi rad/s
df_meh_ppv['RPM'] = df_meh_ppv['RPM'].apply(lambda x: x * (2*np.pi/60))

# Ganti nanometer ke meter
df_meh_ppv['Ketebalan'] = df_meh_ppv['Ketebalan'].apply(lambda x: x/(1e+9))

df_meh_ppv.head()
```

```python
# Define constants
c_0_meh_ppv = 6 # kg/m^3
h_0_meh_ppv = df_meh_ppv['Ketebalan'].iloc[0] # meter
x_meh_ppv = df_meh_ppv['RPM']
y_meh_ppv = df_meh_ppv['Ketebalan']
x_fit_meh_ppv = np.linspace(df_meh_ppv['RPM'].iloc[0], df_meh_ppv['RPM'].iloc[-1], 1000)
```

## **I. P3HT**

### **Model Definition**

```python
def f_ebp(x, k):
  return h_0_p3ht/np.sqrt(1 + k * (h_0_p3ht ** 2) * (x ** 2)) # k = (4 * rho * t)/(3 * eta)

def f_meyerhofer(x, k):
  return np.cbrt(k * ((1 - c_0_p3ht) ** (-1)) * (x ** (-1/2))) # k = (3 * eta_0 * E)/(2 * rho)

def f_empirik(x, k):
  return k * (x ** (-1/2))
```

### **Fitting Model EBP**

```python
param_ebp, param_cov_ebp = curve_fit(f_ebp, x_p3ht, y_p3ht)

print(f"Koefisien: {param_ebp}, Kovarian koefisien: {param_cov_ebp}")
```

```python
ans_ebp = f_ebp(x_fit_p3ht, param_ebp[0])
```

```python
plt.plot(x_p3ht, y_p3ht, 'o', color='red', label='Data')
plt.plot(x_fit_p3ht, ans_ebp, '--', color='blue', label='Fitting')
plt.legend()
plt.xlabel("Kecepatan sudut putar (rad/s)")
plt.ylabel("Ketebalan film (m)")
plt.suptitle("Kurva Ketebalan Lapisan Tipis P3HT", weight="black", x=mid)
plt.title("Fungsi Kecepatan Sudut Spin-Coater")
plt.show()
```

Menentukan nilai $R^2$ dari hasil fitting

```python
# Standard deviation of fit
stddev_ebp = np.sqrt(np.diag(param_cov_ebp))
print(stddev_ebp)
```

### **Fitting Model Meyerhofer**

```python
param, param_cov = curve_fit(f_meyerhofer, x_p3ht, y_p3ht)

print(f"Koefisien: {param}, Kovarian koefisien: {param_cov}")
```

```python
ans_meyerhofer = f_meyerhofer(x_fit_p3ht, param[0])
```

```python
plt.plot(x_p3ht, y_p3ht, 'o', color='red', label='Data')
plt.plot(x_fit_p3ht, ans_meyerhofer, '--', color='blue', label='Fitting')
plt.legend()
plt.xlabel("Kecepatan sudut putar (rad/s)")
plt.ylabel("Ketebalan film (m)")
plt.suptitle("Kurva Ketebalan Lapisan Tipis P3HT", weight="black", x=mid)
plt.title("Fungsi Kecepatan Sudut Spin-Coater")
plt.show()
```

### **Fitting Model Empirik**

```python
param, param_cov = curve_fit(f_empirik, x_p3ht, y_p3ht)

print(f"Koefisien: {param}, Kovarian koefisien: {param_cov}")
```

```python
ans_empirik = f_empirik(x_fit_p3ht, param[0])
```

```python
plt.plot(x_p3ht, y_p3ht, 'o', color='red', label='Data')
plt.plot(x_fit_p3ht, ans_empirik, '--', color='blue', label='Fitting')
plt.legend()
plt.xlabel("Kecepatan sudut putar (rad/s)")
plt.ylabel("Ketebalan film (m)")
plt.suptitle("Kurva Ketebalan Lapisan Tipis P3HT", weight="black", x=mid)
plt.title("Fungsi Kecepatan Sudut Spin-Coater")
```

### **Perbandingan Fitting**

```python
plt.plot(x_p3ht, y_p3ht, 'x', label="Eksperimen")
plt.plot(x_fit_p3ht, ans_empirik, label="Model Empirik")
plt.plot(x_fit_p3ht, ans_ebp, label="Model EBP")
plt.plot(x_fit_p3ht, ans_meyerhofer, label="Model Meyerhofer")

plt.suptitle("Perbandingan Kurva Ketebalan Lapisan Tipis P3HT", weight="black")
plt.title("Model EBP, Meyerhofer, dan Empirik")

plt.xlabel("Kecepatan sudut putar (rad/s)")
plt.ylabel("Ketebalan lapisan tipis P3HT (m)")

plt.legend()

plt.show()
```

## **II. R-Dye**

```python
# Ketebalan (meter)
r_dye = {'RPM': rpm, "Ketebalan": [65, 35, 15, 10, 8]}
df_r_dye = pd.DataFrame(data=r_dye)

# Ganti RPM menjadi rad/s
df_r_dye['RPM'] = df_r_dye['RPM'].apply(lambda x: x * (2*np.pi/60))

# Ganti nanometer ke meter
df_r_dye['Ketebalan'] = df_r_dye['Ketebalan'].apply(lambda x: x/(1e+9))

df_r_dye.head()
```

```python
# Define constants
h_0 = df_r_dye['Ketebalan'].iloc[0] # meter
rho = 2 # kg/m^3
```

```python
x = df_r_dye['RPM']
y = df_r_dye['Ketebalan']
x_fit = np.linspace(df_r_dye['RPM'].iloc[0], df_r_dye['RPM'].iloc[-1], 1000)
```

### **Fitting Model EBP**

```python
# Model EBP
def function(x, a):
  return h_0/(np.sqrt(1 + ((4 * rho * (x**2) * (h_0**2) * t)/(3 * a))))
```

```python
param, param_cov = curve_fit(function, x, y)

print(f"Koefisien: {param}, Kovarian koefisien: {param_cov}")
```

```python
ans = function(x_fit, param[0])
```

```python
plt.plot(x, y, 'o', color='red', label='data')
plt.plot(x_fit, ans, '--', color='blue', label='curve fit')
plt.legend()
plt.show()
```

### **Fitting Model Meyerhofer**

```python
# Model Meyerhofer
def function(x, a, b):
  return np.cbrt((3 * a * b)/(2 * (1-rho) * rho * (x ** 2)))
```

```python
param, param_cov = curve_fit(function, x, y)

print(f"Koefisien: {param}, Kovarian koefisien: {param_cov}")
```

```python
ans = function(x_fit, param[0], param[1])
```

```python
plt.plot(x, y, 'o', color='red', label='data')
plt.plot(x_fit, ans, '--', color='blue', label='curve fit')
plt.legend()
plt.show()
```

### **Fitting Model Empirik**

```python
# Model Empirik
def function(x, a):
  return a * (x ** (-1/2))
```

```python
param, param_cov = curve_fit(function, x, y)

print(f"Koefisien: {param}, Kovarian koefisien: {param_cov}")
```

```python
ans = function(x_fit, param[0])
```

```python
plt.plot(x, y, 'o', color='red', label='data')
plt.plot(x_fit, ans, '--', color='blue', label='curve fit')
plt.legend()
plt.show()
```
